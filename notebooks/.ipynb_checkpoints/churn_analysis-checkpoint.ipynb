```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
from sklearn.metrics import roc_curve

from imblearn.over_sampling import SMOTE

import joblib
```

```python
df = pd.read_csv("../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

```python
print("Dataset Shape:", df.shape)

df.info()
```

```python
# Convert TotalCharges to numeric
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

# Remove missing values
df = df.dropna()

# Remove unnecessary column
df = df.drop("customerID", axis=1)

print("Cleaned dataset shape:", df.shape)
```

```python
df["Churn"] = df["Churn"].map({"Yes":1,"No":0})

df["Churn"].value_counts()
```

```python
import os
os.makedirs("../images", exist_ok=True)

sns.countplot(x="Churn", data=df)

plt.title("Customer Churn Distribution")

plt.savefig("../images/churn_distribution.png")

plt.show()
```

```python
sns.countplot(x="Contract", hue="Churn", data=df)

plt.title("Churn vs Contract Type")

plt.show()
```

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)

plt.title("Monthly Charges vs Churn")

plt.show()
```

```python
df_feat = df.copy()

# Contract risk score
contract_risk = {
    "Month-to-month":3,
    "One year":2,
    "Two year":1
}

df_feat["contract_risk"] = df_feat["Contract"].map(contract_risk)

# High spender feature
df_feat["is_high_spender"] = (df_feat["MonthlyCharges"] > 70).astype(int)

# Tenure risk
df_feat["tenure_risk"] = pd.cut(
    df_feat["tenure"],
    bins=[0,6,12,24,72],
    labels=[3,2,1,0]
).astype(int)

df_feat.head()
```

```python
X = df_feat.drop("Churn", axis=1)

y = df_feat["Churn"]

print(X.shape, y.shape)
```

```python
categorical_cols = X.select_dtypes(include="object").columns

print(categorical_cols)
```

```python
X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

print("New feature shape:", X.shape)
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
smote = SMOTE(random_state=42)

X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("After SMOTE:")
print(y_train_res.value_counts())
```

```python
lr_model = LogisticRegression(max_iter=1000)

lr_model.fit(X_train_res, y_train_res)

lr_pred = lr_model.predict(X_test)

lr_prob = lr_model.predict_proba(X_test)[:,1]
```

```python
print("Logistic Regression Accuracy:", accuracy_score(y_test, lr_pred))

print(classification_report(y_test, lr_pred))
```

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train_res, y_train_res)

rf_pred = rf_model.predict(X_test)

rf_prob = rf_model.predict_proba(X_test)[:,1]
```

```python
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))

print(classification_report(y_test, rf_pred))
```

```python
cm = confusion_matrix(y_test, rf_pred)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

plt.title("Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

```python
fpr, tpr, _ = roc_curve(y_test, rf_prob)

plt.plot(fpr, tpr, label="Random Forest")

plt.plot([0,1],[0,1],"--")

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.legend()

plt.show()

print("ROC AUC:", roc_auc_score(y_test, rf_prob))
```

```python
importance = pd.Series(rf_model.feature_importances_, index=X.columns)

top_features = importance.sort_values(ascending=False).head(10)

top_features.plot(kind="barh")

plt.title("Top Features Influencing Churn")

plt.savefig("../images/feature_importance.png")

plt.show()
```

```python
joblib.dump(rf_model, "../models/churn_model.pkl")

joblib.dump(scaler, "../models/scaler.pkl")

print("Model saved successfully")
```
